```python
# Length Conversion Function
def convert_length(value, from_unit, to_unit):

    if from_unit == 'm' and to_unit == 'ft':
        return value * 3.28084  # 1 meter = 3.28084 feet
    elif from_unit == 'ft' and to_unit == 'm':
        return value * 0.3048   # 1 foot = 0.3048 meters
    else:
        raise ValueError("Only m↔ft conversions supported")
print(convert_length(5, 'm', 'ft'))
print(convert_length(10, 'ft', 'm'))

```

```python
#Weight Conversion Function
def convert_weight(value, from_unit, to_unit):
    """
    Converts weight between kilograms (kg) and pounds (lbs).
    """
    if from_unit == 'kg' and to_unit == 'lbs':
        return value * 2.20462  # 1 kg = 2.20462 lbs
    elif from_unit == 'lbs' and to_unit == 'kg':
        return value * 0.453592 # 1 lb = 0.453592 kg
    else:
        raise ValueError("Only kg↔lbs conversions supported")
print(convert_weight(2.5, 'kg', 'lbs'))
print(convert_weight(10, 'lbs', 'kg'))
```

```python
#Volume Conversion Function
def convert_volume(value, from_unit, to_unit):
    """
    Converts volume between liters (L) and gallons (gal).
    """
    if from_unit == 'L' and to_unit == 'gal':
        return value * 0.264172  # 1 L = 0.264172 gal
    elif from_unit == 'gal' and to_unit == 'L':
        return value * 3.78541   # 1 gal = 3.78541 L
    else:
        raise ValueError("Only L↔gal conversions supported")
print(convert_volume(5, 'L', 'gal'))
print(convert_volume(10, 'gal', 'L'))
```

```python
#Sum function
def calculate_sum(numbers):

    if not numbers:
        raise ValueError("Cannot calculate sum of empty list")
    return sum(numbers)

# Test it:
print(calculate_sum([1, 2, 3, 4, 5]))
print(calculate_sum([10.5, 20.3]))
```

```python
#Average function
def calculate_average(numbers):
    if not numbers:
        raise ValueError("Cannot calculate average of empty list")
        return sum(numbers) / len(numbers)
print(calculate_average([1, 2, 3, 4, 5]))
print(calculate_average([10, 20, 30]))
```

```python
#Maximum function
def find_max(numbers):
    if not numbers:
        raise ValueError("Cannot find maximum of empty list")
    return max(numbers)

# Test it:
print(find_max([1, 5, 3, 9, 2]))
print(find_max([10.5, 2.1, 8.7]))
```

```python
#Minimum Function
def find_minimum(numbers):
    if not numbers:
        raise ValueError("Cannot find minimum of empty list")
    return min(numbers)

# Test it:
print(find_minimum([1, 5, 3, 9, 2]))
print(find_minimum([10.5, 2.1, 8.7]))
```

```python
#Get Numbers Input Function
def get_numbers_from_input():

    user_input = input("Enter numbers separated by spaces: ").strip()
    if not user_input:
        raise ValueError("No numbers provided")

    try:
        numbers = [float(x) for x in user_input.split()]
        return numbers
    except ValueError:
        raise ValueError("All inputs must be valid numbers")
```

```python
#Menu Function
def show_menu():
    print("=== Math Operations on List ===")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")
    print("-" * 20)

def get_user_choice():
    try:
        choice = input("Select operation (1-4): ").strip()
        if choice not in ['1', '2', '3', '4']:
            raise ValueError("Please select 1, 2, 3, or 4")
        return choice
    except ValueError as e:
        print(f"Error: {e}")
        return None
```

```python
# Main Function Structure
def main():
    show_menu()

    choice = get_user_choice()
    if choice is None:
        return

    try:
        numbers = get_numbers_from_input()

        if choice == '1':
            result = calculate_sum(numbers)
            print(f"Sum: {result}")
        elif choice == '2':
            result = calculate_average(numbers)
            print(f"Average: {result:.4f}")
        elif choice == '3':
            result = find_maximum(numbers)
            print(f"Maximum: {result}")
        elif choice == '4':
            result = find_minimum(numbers)
            print(f"Minimum: {result}")

    except ValueError as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```

```python
#Extract Every Other Element
def extract_every_other(lst):

    return lst[::2]
print(extract_every_other([1, 2, 3, 4, 5, 6]))
print(extract_every_other(['a', 'b', 'c', 'd']))
```

```python
#Slice a Sublist
def get_sublist(lst, start, end):

    return lst[start:end+1]

print(get_sublist([1, 2, 3, 4, 5, 6], 2, 4))
print(get_sublist(['a', 'b', 'c', 'd', 'e'], 1, 3))
```

```python
#Reverse a List Using Slicing python
def reverse_list(lst):
    return lst[::-1]
print(reverse_list([1, 2, 3, 4, 5]))
print(reverse_list(['a', 'b', 'c']))
```

```python
# Remove First and Last Elements
def remove_first_last(lst):
   return lst[1:-1]
print(remove_first_last([1, 2, 3, 4, 5]))
print(remove_first_last(['a', 'b', 'c', 'd']))
```

```python
#Get First n Elements
def get_first_n(lst, n):
    return lst[:n]

print(get_first_n([1, 2, 3, 4, 5], 3))
print(get_first_n(['a', 'b', 'c', 'd'], 2))
```

# 4.3 Exercise on Nested List:

```python
#Flatten a Nested List
def flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result
print(flatten([[1, 2], [3, 4], [5]]))
```

```python
# Access Nested List Element
def access_nested_element(lst, indices):
    current = lst
    for i in indices:
        current = current[i]
    return current

nested = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(access_nested_element(nested, [1, 2]))
```

```python
#Sum All Elements in Nested List
def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

# Test it:
print(sum_nested([[1, 2], [3, [4, 5]], 6]))  # Output: 21
```

```python
#Remove Specific Element
def remove_element(lst, elem):

    if isinstance(lst, list):
        return [remove_element(item, elem) for item in lst if item != elem]
    return lst

print(remove_element([[1, 2], [3, 2], [4, 5]], 2))
```

```python
#Find Maximum in Nested List
def find_max(lst):
    if not lst:
        raise ValueError("Cannot find maximum of empty list")

    max_val = float('-inf') # Initialize with negative infinity

    for item in lst:
        if isinstance(item, list):
            current_max = find_max(item)
        else:
            current_max = item

        if current_max > max_val:
            max_val = current_max

    return max_val

print(find_max([[1, 2], [3, [4, 5]], 6]))
```

```python
#Count Occurrences
def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count


print(count_occurrences([[1, 2], [2, 3], [2, 4]], 2))
```

```python
#Deep Flatten
def deep_flatten(lst):

    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

print(deep_flatten([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]))
```

```python
#Nested List Average
def average_nested(lst):

    total = sum_nested(lst)
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_elements(item)
        else:
            count += 1
    return total / count

def count_elements(lst):

    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_elements(item)
        else:
            count += 1
    return count

print(average_nested([[1, 2], [3, 4], [5, 6]]))
```

# PROBLEM 4: Matrix Operations

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
print("A:\n", A)
print("B:\n", B)
```

```python
A_inv = np.linalg.inv(A)
identity = np.dot(A, A_inv)
print("4.1 A·A⁻¹:")
print("Result (should be identity):\n", identity)
print("Is identity?", np.allclose(identity, np.eye(2)))
```

```python
AB = np.dot(A, B)
BA = np.dot(B, A)
print("4.2 Matrix Multiplication:")
print("AB =\n", AB)
print("BA =\n", BA)
print("AB ≠ BA?", not np.array_equal(AB, BA))
```

```python
AB_T = np.transpose(np.dot(A, B))
BT_AT = np.dot(np.transpose(B), np.transpose(A))
print("4.3 Transpose property:")
print("(AB)ᵀ =\n", AB_T)
print("BᵀAᵀ =\n", BT_AT)
print("Equal?", np.allclose(AB_T, BT_AT))
```

# 4.4 Solve Linear System AX = B

```python
# Matrix form: A * X = B
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3,  1, -1]])

B = np.array([-1, -3, 9])

# Method 1: Solve directly
X = np.linalg.solve(A, B)
print("4.4 Linear system solution:")
print("X =", X)
print("x =", X[0], "y =", X[1], "z =", X[2])

# Method 2: Using inverse
X_inv = np.dot(np.linalg.inv(A), B)
print("Using inverse:", X_inv)
```
